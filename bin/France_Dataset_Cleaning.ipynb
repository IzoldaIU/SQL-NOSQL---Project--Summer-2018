```python
def missing_data(data):
    total = data.isnull().sum().sort_values(ascending = False)
    percent = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    return pd.concat([total, percent], axis=1, keys=['Total NaN Values', 'Percentage of NaN Values'])

missing_data(accidents)
```

```python
accidents['Year'] = accidents.Year + 2000
for name in ['Year','Month','Day']:
    accidents[name] = accidents[name].astype('str')
    
accidents['Date'] = accidents.Year + '-' + accidents.Month + '-' + accidents.Day
accidents['Date'] = pd.to_datetime(accidents.Date, format='%Y-%m-%d')
accidents.drop(['Year','Month', 'Day', 'Hour'], axis=1, inplace=True)
```
